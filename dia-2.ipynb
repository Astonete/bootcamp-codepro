Solicitar al usuario que ingrese una frase.

Imprimir un mensaje cada vez que se encuentre una letra "o".

```python
frase_usuario= input("Escribe una frase: ")# se solicita al usuario que ingrese una frase
contador=0# se inicializa un contador en 0 para contar las letras 'o' y 'O'
for frase in frase_usuario:
    if frase == 'o' or frase == 'O' or frase == 'ó' or frase== 'Ó':
        contador= contador+1

print(f'la frase: {frase_usuario} tiene la letra: {frase} unas {contador}')
```
