Challenge 2 🤺 - 15 min
Crear una función que reciba dos letras (parámetros) y que verifique si son iguales:

Si son iguales, imprimir "Las letras son iguales".

De lo contrario, imprimir "Las letras son diferentes".

```python
letraPrimera=input("digite una letra")# ingresa una letra
LetraSegunda=input("digite otra letra")# ingresa otra letra para comparar con la primera letra

def Recibe_letras(primera_letra,segunda_letra):# se define la función que recibe las letras para comparar
    if(primera_letra==segunda_letra):# se compara si las letras son iguales
        print(f"La Letra {primera_letra} es igual a la segunda Letra {segunda_letra}")# se imprime que las letras son iguales
    else:
        print(f"La Letra {primera_letra} Es completamente diferente a la segunda Letra {segunda_letra}")# se imprime que las letras son diferentes

Recibe_letras (letraPrimera,LetraSegunda)# se llama a la función que compara las letras ingresadas por el usuario
```
